```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('swedish_population_by_year_and_sex_1860-2022.csv').drop(['sex'], axis=1)
df['age'] = pd.to_numeric(df['age'], errors='coerce')

children = df[df['age'] < 15].sum()
laborforce = df[(df['age'] > 14) & (df['age'] < 65)].sum()
elder = df[df['age'] > 64].sum()

dependency_ratio = 100 * (children + elder) / laborforce
plt.plot(dependency_ratio.iloc[1:], color='pink')
plt.xlabel('Years')
plt.ylabel('Dependency ratio')
plt.title('Dependency ratio in Sweden from 1860 - 2022')
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::20])
```

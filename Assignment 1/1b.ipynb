```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('swedish_population_by_year_and_sex_1860-2022.csv').drop(['sex'], axis=1)
df['age'] = pd.to_numeric(df['age'], errors='coerce')

children = df[df['age'] < 15].sum()
elder = df[df['age'] > 64].sum()
dependent = children + elder

plt.plot(dependent.iloc[1:], color='red', label='Dependent')
plt.plot(children.iloc[1:], color='pink',label='Children')
plt.plot(elder.iloc[1:], color='orange',label='Elders')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Distribution Over Years')
plt.legend()
ax = plt.gca()
ax.set_xticks(ax.get_xticks()[::20])
plt.show()
```
